### Delete the columns 'Unnamed: 0' and 'Id'

 Are there more male or female names in the dataset?

 Արդյոք dataset-ում ավելի շատ են կանացի թե տղամարդու անունները?

```python
sum_f=0
sum_m=0
a=len(baby_names)

for i in range(a):
    if baby_names["Gender"][i]=="F":
        sum_f+=baby_names["Count"][i]
    else:
        sum_m+=baby_names["Count"][i]

if sum_f==sum_m:
    print("havasar en")
elif sum_f>sum_m:
    print("kanaci anunnery aveli shat en")
else:
    print("txamardkanc anunnery aveli shaten")
```
